```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_csv("virat-kohli.csv")
```

```python
# Display the dataset.
data.head()
```

```python
data['Year'] = data['Year'].astype(int)
data['Runs'] = data['Runs'].astype(int)
data['Avg'] = data['Avg'].astype(float)
```

```python
# Extract relevant columns
years = data['Year']
runs = data['Runs']
average = data['Avg']
```

```python
# Create figure and subplots using OOP approach
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), constrained_layout=True)
fig.suptitle("Virat Kohli's IPL Performance Trends (2008-2025)", fontsize=16)

# First subplot: Bar plot for Runs
ax1.bar(years, runs, color='royalblue')
ax1.set_title("Runs Scored Over Years")
ax1.set_xlabel("Year")
ax1.set_ylabel("Runs Scored")
ax1.grid(True, linestyle='--', alpha=0.5)

# Second subplot: Line plot for Batting Average
ax2.plot(years, average, marker='o', color='darkorange', linewidth=2)
ax2.set_title("Batting Average Over Years")
ax2.set_xlabel("Year")
ax2.set_ylabel("Batting Average")
ax2.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
