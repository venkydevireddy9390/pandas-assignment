#### Part:3 Data Cleaning

```python
# Task 1: Remove the S N column from the DataFrame.
remove_SN=data.drop(columns=["S N"],inplace=True)
```

```python
# Task 2: Fill missing values in Inches column using mode.
fill_inches=data["Inches"].fillna(data["Inches"].mode()[0],inplace=True)
```

```python
# Task 3: Fill missing values in Ram column using mode.
fill_Ram=data["Ram"].fillna(data["Ram"].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
#Task 4: Verify missing values handling using df.info() 
data.info()
```

```python
# Task 5: Convert Inches column to numeric type. 
data["Inches"]=pd.to_numeric(data["Inches"].replace(r"?",regex=True))
```

```python
#Task 6: Convert Ram column to numeric by removing text (GB).
data["Ram"]=pd.to_numeric(data["Ram"].astype(str).str.replace(r"[^0-9.]","",regex=True))
```

```python
# Task 7: Convert Weight column to numeric by removing text (kg) 
data["Weight"]=pd.to_numeric(data["Weight"].astype(str).str.replace(r"[^0-9.]","",regex=True))
```

#### Part:5 Creating New Columns

```python
# Task 1: Create a new column ‘Price in $’ by converting Price column into $ Dollars.
usd_rate=83
data["Price in $"]=(data["Price"]/usd_rate)
```

```python
#Task 2: Create Weight_Category column based on laptop weight. 
#“Light” → Weight < 2kg 
#“Heavy” → Weight ≥ 2kg 
import numpy as np
data["Weight_Category"]=np.where(data["Weight"]< 2,  "Light","Heavy")
```
