```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\sekhar\Desktop\AI-FROM-SCRATCH\sales_data.csv")
```

```python
df.head()
```

```python

df.info()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x="Unit_Price",
    y="Customer_Rating",
    hue="Gender"
)

plt.title("Unit Price vs Customer Rating based on Gender")
plt.xlabel("Unit Price")
plt.ylabel("Customer Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x="Sales_Amount",
    kde=True,
    bins=20
)

plt.title("Distribution of Sales Amount with KDE")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```
